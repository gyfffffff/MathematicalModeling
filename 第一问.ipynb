第一问要求分析会员消费特征。将会员按照年龄和性别分类：女性45岁以上，女性45岁以下，男性45岁以上，男性45岁以下。

```python
# 四个dataframe分别是：女性45岁以上，女性45岁以下，男性45岁以上，男性45岁以下
df11 = df1[(df1['xb']==0.0) & (df1['nl']>=45)]
df12 = df1[(df1['xb']==0.0) & (df1['nl']<45)]
df13 = df1[(df1['xb']==1.0) & (df1['nl']>=45)]
df14 = df1[(df1['xb']==1.0) & (df1['nl']<45)]
```

(1) 分析该商场会员的消费特征，比较会员与非会员群体的差异，并说明会员群体给商场带来的价值。
**通过以上分析，可以看出会员有以下消费特征：**
- 时间上，分布较均匀，全年每天都有人购物，换季时略多，周末略多，下午3点左右和晚上20点左右略多，23点后很少有会员购物。
- 付款金额方面，一次购买在1000-2000之间居多,45岁以上的会员消费更高一些，购买数量经常大于1件，且退货率较低。
- 从以上两方面可以看出会员购买的忠实度较高。
**与会员相比，非会员有着明显的不同：**
- 时间上，非会员几乎只在周日购物，且购物时间几乎在23点，全年之中绝大多数非会员消费集中在年初。
- 付款金额方面，方差较大，均值较小，偏度为负数。（此处可以将会员和非会员数据列一个统计表）
- 从以上两点可以看出非会员忠实度低。
**经过以上分析，会员群体可以给商场带来以下价值：**
- 主要的购买力。与非会员相比，会员消费时间并没有明显的偏态分布，较为均匀，可以给商场提供源源不断的购买力。
- 持续收入。会员的付款均值大于非会员的均值，是商场收入的主要来源。
- 口碑。会员的忠诚度可以成为商场隐形的宣传力量，提高商场的知名度，影响力和认可度。
